```python
def autopct_format(values):
        def my_format(pct):
            total = sum(values)
            val = int(round(pct*total/100.0))
            return '{:.1f}%\n({v:d})'.format(pct, v=val)
        return my_format
    
pal = sns.color_palette("Set2")
pal2 = sns.color_palette("pastel")
```

```python
fig, ax = plt.subplots(2,5, figsize=(20, 10))

# ax[0][0]
h = data['HeartDisease'].value_counts()
ax[0][0].set_title('HeartDisease', y=1)
HeartDisease_pie = ax[0][0].pie(h, labels = h.index, autopct=autopct_format(h), colors = pal2)

# ax[0][1]
s = data['Smoking'].value_counts()
ax[0][1].set_title('Smoking', y=1)
Smoking_pie = ax[0][1].pie(s, labels = s.index, autopct=autopct_format(s), colors = pal)

# ax[0][2]
a = data['AlcoholDrinking'].value_counts()
ax[0][2].set_title('AlcoholDrinking', y=1)
AlcoholDrinking_pie = ax[0][2].pie(a, labels = a.index, autopct=autopct_format(a), colors = pal)

# ax[0][3]
st = data['Stroke'].value_counts()
ax[0][3].set_title('Stroke', y=1)
Stroke_pie = ax[0][3].pie(st, labels = st.index, autopct=autopct_format(st), colors = pal)

# ax[0][4]
d = data['DiffWalking'].value_counts()
ax[0][4].set_title('DiffWalking', y=1)
DiffWalking_pie = ax[0][4].pie(d, labels = d.index, autopct=autopct_format(d), colors = pal)

# ax[1][0]
se = data['Sex'].value_counts()
ax[1][0].set_title('Sex', y=1)
Sex_pie = ax[1][0].pie(se, labels = se.index, autopct=autopct_format(se), colors = pal)

# ax[1][1]
p = data['PhysicalActivity'].value_counts()
ax[1][1].set_title('PhysicalActivity', y=1)
PhysicalActivity_pie = ax[1][1].pie(p, labels = p.index, autopct=autopct_format(p), colors = pal)

# ax[1][2]Asthma
ast = data['Asthma'].value_counts()
ax[1][2].set_title('Asthma', y=1)
Asthma_pie = ax[1][2].pie(ast, labels = ast.index, autopct=autopct_format(ast), colors = pal)

# ax[1][3]
k = data['KidneyDisease'].value_counts()
ax[1][3].set_title('KidneyDisease', y=1)
KidneyDisease_pie = ax[1][3].pie(k, labels = k.index, autopct=autopct_format(k), colors = pal)

# ax[1][4]
sk = data['SkinCancer'].value_counts()
ax[1][4].set_title('SkinCancer', y=1)
SkinCancer_pie = ax[1][4].pie(sk, labels = sk.index, autopct=autopct_format(sk), colors = pal)

plt.show()
```

```python
df=data.copy()
```

```python
df
```

```python
from sklearn import preprocessing

le = preprocessing.LabelEncoder()
for col in df[['Smoking','AlcoholDrinking','Stroke', 'DiffWalking', 'Sex', 'PhysicalActivity', 'Asthma', 'KidneyDisease', 'SkinCancer', 'AgeCategory', 'Race', 'Diabetic', 'GenHealth', 'HeartDisease']]:
    df[col]=le.fit_transform(df[col])
```

```python
df
```

```python
corr = df.corr()
plt.figure(figsize=(20, 9))
k = 18 #number of variables for heatmap
cols = corr.nlargest(k, 'HeartDisease')['HeartDisease'].index
cm = np.corrcoef(df[cols].values.T)
sns.set(font_scale=1.25)
hm = sns.heatmap(cm, cbar=True, annot=True, square=True, fmt='.2f', annot_kws={'size': 10}, yticklabels=cols.values, xticklabels=cols.values)
plt.show()
```

```python
X = df.iloc[:, :-1].values
y = df.iloc[:, -1].values
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(solver='liblinear')
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test)
```

```python
y_pred_proba=classifier.predict_proba(X_test)
```

```python
print(classifier.fit(X_train, y_train).intercept_)
print(classifier.fit(X_train, y_train).coef_)
```

```python
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score, confusion_matrix

print("accuracy_score : {:.2f} %".format(accuracy_score(y_test, y_pred)*100))
print("f1_score: {:.2f} %".format(f1_score(y_test, y_pred)*100))
print("recall_score: {:.2f} %".format(recall_score(y_test, y_pred)*100))
print("precision_score: {:.2f} % ".format(precision_score(y_test,y_pred)*100))
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, ConfusionMatrixDisplay
cm = confusion_matrix(y_test, y_pred)
```

```python
disp = ConfusionMatrixDisplay(confusion_matrix=cm)
disp.plot()
plt.show()
```
